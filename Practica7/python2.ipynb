```
def get_intersection_greater_than_t(set_a, set_b, t):
    result = set()
    for elem in set_a:
        if elem > t and elem in set_b:
            result.add(elem)
    return result

def print_set(s):
    if not s:
        print("Множество пусто")
    else:
        print(" ".join(map(str, sorted(s))))

def main():
    t = int(input("Введите значение t: "))

    print("Введите количество элементов в множестве A: ")
    n = int(input())
    print("Введите элементы множества A:")
    A = set()
    for _ in range(n):
        A.add(int(input()))

    print("Введите количество элементов в множестве B: ")
    n = int(input())
    print("Введите элементы множества B:")
    B = set()
    for _ in range(n):
        B.add(int(input()))
    a_ptr = A
    b_ptr = B

    result = get_intersection_greater_than_t(a_ptr, b_ptr, t)

    print(f"Элементы, присутствующие в обоих множествах и большие {t}: ", end="")
    print_set(result)

if __name__ == "__main__":
    main()
```
