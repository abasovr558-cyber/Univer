```
def count_elements_greater_than_t(arr, t):
    count = 0
    for elem in arr:
        if elem > t:
            count += 1
    return count

def compare_and_print_arrays(arr_a, arr_b, t):
    count_a = count_elements_greater_than_t(arr_a, t)
    count_b = count_elements_greater_than_t(arr_b, t)

    a_ptr = arr_a
    b_ptr = arr_b

    if count_a <= count_b:
        print(f"Массив A имеет наименьшее количество элементов (> t): {count_a}")
        print("Элементы массива A:", *a_ptr)
    else:
        print(f"Массив B имеет наименьшее количество элементов (> t): {count_b}")
        print("Элементы массива B:", *b_ptr)

def main():
    N = 5

    t = int(input("Введите значение t: "))

    print(f"Введите {N} элементов для массива A:")
    A = []
    for i in range(N):
        A.append(int(input()))

    print(f"Введите {N} элементов для массива B:")
    B = []
    for i in range(N):
        B.append(int(input()))

    compare_and_print_arrays(A, B, t)

if __name__ == "__main__":
    main()
```
